```python
import os

# 경로 지정
# 예: Colab 노트북이 있는 디렉토리 또는 Google 드라이브 내의 경로
directory = '/content/drive/MyDrive/대출 데이터 모음'  # 이 경로는 Colab에 기본적으로 있는 샘플 데이터 디렉토리입니다.

# 해당 경로에 있는 파일과 디렉토리의 이름을 모두 나열
files_in_directory = os.listdir(directory)

# 파일 이름 출력
for filename in files_in_directory:
    print(filename)
```
