```python
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer

# KoNLPy의 Okt 객체를 생성
okt = Okt()

# 텍스트 데이터
text = "여기에 텍스트 데이터를 입력하세요."

# Okt를 사용하여 명사 추출 함수 정의
def get_nouns(text):
    nouns = okt.nouns(text)
    return [noun for noun in nouns if len(noun) > 1]  # 한 글자 명사 제외

# TF-IDF 벡터라이저 생성
tfidf_vectorizer = TfidfVectorizer(tokenizer=get_nouns)

# 텍스트 데이터를 기반으로 TF-IDF 학습 및 변환
tfidf_matrix = tfidf_vectorizer.fit_transform([text])

# 각 키워드의 중요도에 따라 정렬
sorted_keywords = sorted(tfidf_vectorizer.vocabulary_.items(), key=lambda x: x[1], reverse=True)

# 중요도가 높은 상위 5개의 키워드 추출
top_keywords = sorted_keywords[:5]

# 결과 출력
print(top_keywords)
```
